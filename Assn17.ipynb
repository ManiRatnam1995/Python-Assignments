1. Assign the value 7 to the variable guess_me. Then, write the conditional tests (if, else, and elif) to
print the string &#39;too low&#39; if guess_me is less than 7, &#39;too high&#39; if greater than 7, and &#39;just right&#39; if equal
to 7.

```python
def guessing():
    print("Welcome to guessing game")
    guess_me=int(input("Enter the number and it must not be a floating point;"))
    if guess_me > 7:
        print("too high")
    elif guess_me == 7:
        print("just right")
    else:
        print("too low")
guessing()
```

```python
def guess_in_loops():
    print("You only go when you guess the correct one")
    guess_me = 7
    variable_start=1
    print("start with ",variable_start)
    while variable_start < guess_me:
        variable_start=int(input("Enter the numnber and it must not be a floating point:"))
        if variable_start > guess_me:
            print("oops")
        elif variable_start == guess_me:
            print("found it")
        else:
            print("too low")
            
guess_in_loops()
```

7. Use a generator comprehension to return the string &#39;Got &#39; and a number for the numbers in
range(10). Iterate through this by using a for loop.

```python
def gen():
    l=[]
    s="Got"
    for char in s:
        l.append(char)
    for number in range(10):
        l.append(number)
    for val in l:
        yield(val)
        
    
    
```

```python
for var in gen():
    print(var,end=" ")
```

9. Define a generator function called get_odds that returns the odd numbers from range(10). Use a
for loop to find and print the third value returned.

```python
def get_odds():
    l=[]
    for char in range(10):
        l.append(char)
    yield(l[2])
```

```python
for char in get_odds():
    print(char)
```
